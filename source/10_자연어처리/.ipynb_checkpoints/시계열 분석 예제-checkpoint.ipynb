```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# 한글, 축의 마이너스 기호 깨짐 방지
plt.rc("font", family = 'Malgun Gothic')
plt.rc('axes', unicode_minus = False)
```

```python
df = pd.read_csv('data/일별평균대기오염도_2022(에어코리아).csv', encoding = 'cp949')
df.head()
```

```python
df.info()
```

```python
# 측정일시 컬럼을 날짜형으로 변환
# Sol 1)
# int64 → str 형변환
df['측정일시'] = df['측정일시'].astype('str')
# "연도-월-일" 형식으로 변경
df['측정일시'] = df['측정일시'].str[:4] + '-' +\
                df['측정일시'].str[4:6] + '-' +\
                df['측정일시'].str[6:]
display(df['측정일시'])
display(df.head(1))
```

```python
# Sol 2)
df['측정일시'] = pd.to_datetime(df['측정일시'])
df.info()
```

```python
df['측정소명'].unique()
```

```python
df_filter = df.loc[df['측정소명'] == '공항대로']
df_filter.head()
```

```python
plt.figure(figsize = (20, 8))
plt.plot(df_filter['측정일시'], df_filter['미세먼지농도(㎍/㎥)'], label = '미세먼지')
plt.plot(df_filter['측정일시'], df_filter['초미세먼지농도(㎍/㎥)'], label = '초미세먼지')
plt.legend(loc = 'upper left')
plt.show()
# 미세먼지 데이터가 변화 더 잘 관찰될 것으로 예상됨
```

```python
# 측정일시, 미세먼지농도 데이터만 추출
df_filter2 = df_filter[['측정일시', '미세먼지농도(㎍/㎥)',]]
df_filter2.columns = ['측정일시', '미세먼지농도']
df_filter3 = df_filter2.set_index('측정일시') # Series로 변경
df_filter3.sample(5)
```

```python
# observed(실제값), trend(추세), seasonal(계절성), residual(잔차)
result = seasonal_decompose(df_filter3['미세먼지농도'],
                            period = 30,        # 30일 단위로 계절성을 분석
                            model = 'additive', # 계절성, 추세, 잔차를 덧셈으로 분석 
                           )
fig, axes = plt.subplots(4, 1, # 3행 1열
                         figsize = (12, 8))
result.observed.plot(ax = axes[0])
axes[0].set_ylabel('실제값')
result.trend.plot(ax = axes[1])
axes[1].set_ylabel('trend')
result.seasonal.plot(ax = axes[2])
axes[2].set_ylabel('계절성')
result.resid.plot(ax = axes[3])
axes[3].set_ylabel('잔차')
plt.show()
```

```python
# facebook에서 개발한 시계열 예측 라이브러리 
# pip install prophet
from prophet import Prophet

# 컬럼명 변경
df_filter2.columns = ['ds', 'y']
```

```python
p = Prophet()
p.fit(df_filter2)
```

```python
# 30일 이후의 데이터를 예측
future = p.make_future_dataframe(periods = 30) 
forecast = p.predict(future)
f = p.plot(forecast)
```

```python
# trend와 휴일효과 그래프
f2 = p.plot_components(forecast)
```

```python
# ds: 원데이터, 날짜
# yhat: 예측값
# yhat_lower: 예측의 최소 범위
# yhat_upper: 예측의 최대 범위
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)
```
