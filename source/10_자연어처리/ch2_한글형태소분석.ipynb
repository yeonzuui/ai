## 3.1 HanNanum

```python
from konlpy.tag import Hannanum
hannanum = Hannanum(jvmpath = None,
                    max_heap_size = 1024)
# jvmpath: 자바 가상머신 경로
# max_heap_size: 최대 힙메모리 사이즈(단위 Byte)
# 기본값: 1024
# '1g', '2048m' 식으로도 표현 가능
hannanum.analyze(text)
```

### Pos Tagging Chart 다운로드
* knolpy docs 구글링 후 https://konlpy.org/ko/latest/
    * API 메뉴 클릭 → 품사 태깅 클래스 간 비교(Comparision between POS tagging classes)
    * "한국어 품사 태그 비교표(Korean POS tags comparison chart)
    * ntags: 태그 종류 개수

```python
# 형태소 분석: morphs
print(hannanum.morphs(text))
```

```python
# 명사만 추출: nouns
print(hannanum.nouns(text))
```

```python
# 품사 태그: pos
print(hannanum.pos(text))
# 기본값: ntags = 9
```

```python
print(hannanum.pos(text, ntags = 22))
```

```python
# Quiz1: text중에서 형용사(PA)만 추출
tagged_list = hannanum.pos(text, ntags = 22)

# sol 1
[token for token, tag in tagged_list if tag == 'PA']

# sol 1-1
[tagtext[0] for tagtext in tagged_list if tagtext[1] == 'PA']

# sol 2
pa_set = set()
for token, tag in tagged_list:
    if tag == 'PA':
        pa_set.add(token)
print(pa_set)
```

```python
# Quiz2: text중에서 일반명사(NC)만 추출
tagged_list = hannanum.pos(text, ntags = 22)

# sol 1
[word for word, tag in tagged_list if tag == 'NC']

# sol 1-1
[tagtext[0] for tagtext in tagged_list if tagtext[1] == 'NC']

# sol 2
pa_set = set()
for token, tag in tagged_list:
    if tag == 'NC':
        pa_set.add(token)
print(pa_set)
```

## 3.2 Kkma

```python
from konlpy.tag import Kkma
kkma = Kkma()
# max_heap_size: 최대 힙메모리 사이즈(단위 Byte), 기본값 1024, '1g', '2048m' 가능

# 형태소 분석
print(kkma.morphs(text))
```

```python
# 명사 추출
print(kkma.nouns(text)) 
```

```python
# pos tagging
print(kkma.pos(text))
# ntags 지정 불가, 56개 고정
```

```python
# 보통명사(NNG), 고유명사(NNP)
tagged_text = kkma.pos(text)
[word for word, tag in tagged_text if tag == 'NNG' or tag == 'NNP']    # sol 1) or
[word for word, tag in tagged_text if (tag == 'NNG') | (tag == 'NNP')] # sol 2) 하이픈
[word for word, tag in tagged_text if tag in('NNG', 'NNP')]            # sol 3) in연산자
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt # plt.imshow()
```

```python
# 영어로 wordcloud
# 1) emma 객체 생성 → RegexpTokenizer(정규표현식 이용) → pos tagging
import nltk
from nltk.tokenize import RegexpTokenizer # 단어 추출
from nltk.tag import pos_tag
# 'D:\nltk_data\corpora\gutenberg\austen-emma.txt'
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')

# 단어 추출
# \w: word(문자), {3,}: 3개 이상
ret = RegexpTokenizer('[\w]{3,}')
words = ret.tokenize(emma)

# 명사만 추출
emma_tags = pos_tag(words)
# noun_list = [emma_tag[0] for emma_tag in emma_tags if emma_tag[1] in ('NN', 'NNS')]
noun_list = [word for word, tag in emma_tags if tag in ('NN', 'NNS')]
emma_noun = ' '.join(noun_list)
emma_noun[:100]
```

```python
# 명사 개수
len(noun_list), len(set(noun_list)) # 중복 제거
```

```python
# 단어의 빈도수 시각화(nltk.Text)
# 워드클라우드에서는 출현된 명사들을 ' '로 연결한 문자: data_noun - 대한민국, 헌법, 유구, ...
# nltk.Text().plot(출현한 명사들 list): word_list ['대한민국', '헌법', '유구', ...]
import nltk
import matplotlib.pyplot as plt
from konlpy.tag import Komoran
from konlpy.corpus import kolaw
data = kolaw.open('constitution.txt').read() # 한 줄씩 list로 가져옴
komoran = Komoran()

# 일반명사(NNG), 고유 명사(NNP), 의존 명사(NNB)
word_list = [token for token, tag in komoran.pos(data) if tag in ('NNG', 'NNP', 'NNB')]
data_text = nltk.Text(word_list)
# 한글 깨짐 방지 설정
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (15, 4))
data_text.plot(15)
plt.show()
```

```python
# 튜플 매개변수 - list로 감싸기
words, counts = zip(*data_text.vocab().most_common(15))
plt.figure(figsize = (15, 4))
plt.plot(words, counts)
plt.xticks(rotation = 10) # rotation 기본값: 0
plt.xlabel("Samples")
plt.ylabel("Counts")
plt.grid()
plt.show()
```
