```python
import pandas as pd
import numpy as np
```

```python
# Data read
# 어획량 Data
fish_catch = pd.read_csv('data/어획량_수정본.csv')
# 방류 Data
fish_discharge = pd.read_csv('data/방류실적_수정.csv')
```

```python
display(fish_catch, fish_discharge)
```

```python
print(fish_catch['어종명'].unique(), fish_discharge['품종'].unique())
print(len(fish_catch['어종명'].unique()), '종,', len(fish_discharge['품종'].unique()), '종')
```

```python
fish_catch_species = fish_catch['어종명'].unique()
fish_discharge_varieties = fish_discharge['품종'].unique()

species_to_varieties = {
    ['가오리류', '갈치', '강달이류', '고등어', '기름가자미', '기타다랑어류', '꽁치', '눈볼대', '망둑어류', '멸치', ] : '기타',
    '가자미류' : ['돌가자미', '문치가자미', '강도다리', '도다리'],
    '감성돔' : ['감성돔'],
    '기타돔류' : ['자바리', '붉바리'],
    '넙치류(광어)' : ['넙치'],
    '농어류' : ['점농어'],
    '능성어' : ['능성어'],
    '대구' : ['대구'],
    '돌돔(줄돔)' : ['돌돔'],
    '민어' : ['민어'],
    '반지' : ['기타'],
    '방어류' : ['기타'],
    '밴댕이류' : ['기타'],
    '병어류' : ['기타'],
    '보구치' : ['기타'],
    '복어류' : ['황복'],
    '옥돔류' : ['기타'],
    '자리돔' : ['기타'],
    '참다랑어' : ['기타'],
    '참돔' : ['참돔'],
    '학공치' : ['기타'],
    '기타류' : ['갑오징어', '개볼락', '꺽지', '꼼치', '낙지', '명태', '박대', '볼락', '연어', '왕우럭', '은어',
            '주꾸미', '쥐노래미', '쥐치', '참조기', '황어', '대문어', '전어']
    
 }

# 데이터 프레임으로 변경
df_catch = pd.DataFrame({'어종명': fish_catch_species})
df_discharge = pd.DataFrame({'품종': fish_discharge_varieties})

# 매핑 정보를 활용해 데이터 통합
def map_species(row, mapping):
    for species, varieties in mapping.items():
        if row in varieties:
            return species
    return '기타'

df_discharge['어종명'] = df_discharge['품종'].apply(lambda x: map_species(x, species_to_varieties))

print(df_discharge['품종'].unique())
print(df_discharge['어종명'].unique())
display(df_discharge.sample(15))
display(df_catch.sample(5))
```

```python
# 품종 고유값 추출
unique_varieties = df_discharge['품종'].unique()

# 매핑에 포함되지 않은 품종 찾기
unmapped_varieties = set(unique_varieties) - set([item for sublist in species_to_varieties.values() for item in sublist])
print("매핑되지 않은 품종:", unmapped_varieties)
```
