# 2절. K-Means 클러스터링
- 가장 단순하고 빠른 클러스터링 알고리즘
- 몇 개 클러스터로 나눌지 미리 지정해야 함

```python
# 그래프 저장시 해상도 높게
%config InlineBackend.figure_format = 'retina'
# 패키지 import 
import matplotlib.pyplot as plt
import seaborn as sns
# 한글설정
plt.rc('font', family = 'Malgun Gothic') # 윈도우즈
plt.rc('axes', unicode_minus = False) # 축의 - 깨짐 방지
```

```python
# 분류를 위한 가상 데이터셋 생성
from sklearn.datasets import make_classification

X, y = make_classification(n_samples            = 20, # 데이터 개수(기본값 100)
                           n_features           = 2,  # 독립변수 개수(기본값 20)
                           n_informative        = 2,  # 종속변수에 영향을 미치는 변수의 수
                           n_redundant          = 0,  # 종속변수에 영향을 미치지 않는 변수의 수
                           n_clusters_per_class = 1,  # 각 클래스당 군의 수(1이면 군에 서브 그룹이 없다)
                           n_classes            = 2,  # 종속변수의 클래스 수
                           random_state         = 123)
X.shape, y.shape
```

```python
import numpy as np
np.c_[X, y][:3]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 3))
plt.scatter(x = X[:, 0],
            y = X[:, 1],
            c = y, s = 100)
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2,  # 클러스터의 개수
               init = 'random', # random(중심초기점이 random), k-means++(멀리 떨어진 초기점)
               n_init = 4,
               max_iter = 300)
model.fit(X)
```

```python
model.cluster_centers_ # 최종 두 클러스터의 중심점
```

```python
pred = model.predict(X)
print("예 측 값  :", pred)
print("modelLabel: ", model.labels_)
print('실제 y값:', y)
```

```python
# k-means가 0 그룹분류된 X데이터
X[model.labels_ == 0]
```

```python
c0, c1 = model.cluster_centers_ # 중심점 2개
plt.figure(figsize = (12, 3))
plt.scatter(x = X[:, 0], 
            y = X[:, 1],
            marker = 'x', 
            c = y, s = 200, linewidth = 3)
# 0 그룹 산점도
plt.scatter(x = X[model.labels_ == 0, 0], 
            y = X[model.labels_ == 0, 1], marker = "o", c = "b", s = 100)
# 1 그룹 산점도
plt.scatter(x = X[model.labels_ == 1, 0], 
            y = X[model.labels_ == 1, 1], marker = "o", c = "g", s = 100)
# 0 그룹 중심점
plt.scatter(x = c0[0],
            y = c0[1], marker = 's', c = 'r', s = 150)
# 1 그룹 중심점
plt.scatter(x = c1[0],
            y = c1[1], marker = 's', c = 'cyan', s = 150)
```

```python
import pandas as pd
pd.crosstab(y, model.labels_, rownames = ['실제값'], 
                              colnames = ['k-means 값'])
```

```python
# 거리 계산 시 스케일 조정
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
Xn     = scaler.fit_transform(X)
X_df   = pd.DataFrame(X, columns = ['x0', 'x1'])
Xn_df  = pd.DataFrame(Xn, columns = ['x0_nor', 'x1_nor'])
pd.concat([X_df, Xn_df], axis = 1)
```

```python
model = KMeans(n_clusters = 2, 
               init       = 'k-means++',
               n_init     = 5, 
               max_iter   = 300)
model.fit(Xn)
print('그룹 결과: ', model.labels_)
```

```python
c0, c1 = model.cluster_centers_ # 중심점 2개
plt.figure(figsize = (12, 3))
plt.scatter(x = Xn[:, 0], 
            y = Xn[:, 1],
            marker = 'x', 
            c = y, s = 200, linewidth = 3)
# 0 그룹 산점도
plt.scatter(x = Xn[model.labels_ == 0, 0], 
            y = Xn[model.labels_ == 0, 1], marker = "o", c = "b", s = 100)
# 1 그룹 산점도
plt.scatter(x = Xn[model.labels_ == 1, 0], 
            y = Xn[model.labels_ == 1, 1], marker = "o", c = "g", s = 100)
# 0 그룹 중심점
plt.scatter(x = c0[0],
            y = c0[1], marker = 's', c = 'r', s = 150)
# 1 그룹 중심점
plt.scatter(x = c1[0],
            y = c1[1], marker = 's', c = 'cyan', s = 150)
plt.show()
```

```python
# 회차별 군집 확인하기
# 학습된 모델과 데이터를 매개변수로 받는 plot_cluster 함수
def plot_cluster(model, data):
    c0, c1 = model.cluster_centers_ # 중심점 2개
    # 0 그룹 산점도
    plt.scatter(x = data[model.labels_ == 0, 0], 
                y = data[model.labels_ == 0, 1], marker = "o", c = "b", s = 100)
    # 1 그룹 산점도
    plt.scatter(x = data[model.labels_ == 1, 0], 
                y = data[model.labels_ == 1, 1], marker = "o", c = "g", s = 100)
    # 0 그룹 중심점
    plt.scatter(x = c0[0],
                y = c0[1], marker = 's', c = 'r', s = 150)
    # 1 그룹 중심점
    plt.scatter(x = c1[0],
                y = c1[1], marker = 's', c = 'cyan', s = 150)
    return c0, c1
```

```python
plt.figure(figsize = (8, 8))
# 학습 6번 반복
for i in range(1, 7):
    model = KMeans(n_clusters   = 2, 
                   init         = 'k-means++',
                   n_init       = 10, # 랜덤 초기점 조정 횟수
                   max_iter     = i, 
                   random_state = 1) # seed값
    model.fit(X)
    plt.subplot(3, 2, i)
    c0, c1 = plot_cluster(model, X)
    print("c0({:.2f}, {:.2f}), c1({:.2f}, {:.2f})".format(c0[0], c0[1], c1[0], c1[1]))
plt.show()
```

# 3절. Hierachical 클러스터링 (계층적 군집분석)
- 군끼리 묶으면서 최종적으로는 하나의 군집이 될 때까지 클러스터링하는 알고리즘
- `[ dendrogram ]`이라는 그래프를 이용하여 시각화
- 군의 개수를 미리 정하지 않음

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.sample()
```

```python
# dendrogram이 149개의 그룹에서 시작하여 유사한 데이터끼리 군집화함으로써 범위를 좁혀 감
# 계층적 군집분석을 위해 모든 데이터를 수치화(species 라벨 인코딩)
# 계층적 군집분석은 독립/종속 변수를 구분하지 않으며 데이터에 문자 포함을 허용하지 않음
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris.species)
iris.iloc[::50]
```

```python
from scipy.cluster.hierarchy import linkage
cluster_model = linkage(iris, method = "complete")
# method = 'complete': 두 그룹의 특성 차이를 거리로 측정
```

```python
# 형성된 계층적 모델로 dendrogram
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize = (20, 3))
dendrogram(cluster_model, labels = iris.index)
# 그림 파일로 저장
plt.savefig('ch2_dendrogram.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
# 지정된 임계값으로 그룹화
# 임계값이 4일 때 그룹 3개
from scipy.cluster.hierarchy import fcluster
fcluster(cluster_model,
         t = 4, # 임계값
         criterion = 'distance')
```

```python
# 임계값이 6일 때 그룹 2개
from scipy.cluster.hierarchy import fcluster
fcluster(cluster_model,
         t = 6,
         criterion = 'distance')
```

# 4절. DBSCAN 클러스터링
- 연결 기반 군집화 알고리즘
- 클러스터 수를 미리 정할 필요가 없음
- 이상치를 자동 감지
- eps(반경)과 min_samples(최소 샘플 수)를 적절히 조정해야 함

```python
# rand(size): 균등분포 size개 난수(0 <= x < 1) 발생
noise  = np.random.normal(0, 0.2, X1.shape[0])

X1     = np.random.rand(200)*3
Y1     = np.sin(X1) + noise
group1 = np.c_[X1, Y1, np.zeros(X1.shape[0])]

X2     = X1 + 1.5
Y2     = np.cos(X2) + noise
group2 = np.c_[X2, Y2, np.ones(X2.shape[0])]

data = np.r_[group1, group2] # 400행
plt.scatter(data[:, 0], data[:, 1], c = data[:, 2])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
# 반경 0.3 이내의 포인트들을 동일 그룹으로 묶음
db_model = DBSCAN(eps = .3, min_samples = 10)
db_model.fit(data[:, 0:2])
```

```python
labels = db_model.labels_
# labels가 -1은 이상치
len(set(labels))
```

```python
import matplotlib.pyplot as plt
colors = ['r', 'g']
for i in range(2):
    plt.scatter(x = data[labels == i, 0],
                y = data[labels == i, 1],
                c = colors[i])
    plt.scatter(x = data[labels == -1, 0],
                y = data[labels == -1, 1],
                s = 100, marker = 'x', c = 'k')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris_X = iris.iloc[:, :-1 ]
iris_y = iris.iloc[:, -1] # setosa, viginica, ...
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris_y_enc = le.fit_transform(iris_y)
iris_y[::50], iris_y_enc[::50]
```

```python
# iris_X: 독립변수
# iris_y: 종속변수. 종별 이름(문자)
# iris_y_enc: 종별이름을 인코딩 0, 1...
```

## 5.2 실루엣 계수
- 클러스터 소속도 모르고, 클러스터 개수도 모를 때 평가하는 방법 

```python
from sklearn.metrics import silhouette_score
silhouette_score(iris_X, model3.labels_)
# 단위 % 환산하면 안 됨
```

```python
X1     = np.random.rand(200)*3
Y1     = np.sin(X1) + noise
group1 = np.c_[X1, Y1]

X2     = X1 + 1.5
Y2     = np.cos(X2) + noise
group2 = np.c_[X2, Y2]

data = np.r_[group1, group2] # 400행
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
# 실루엣 계수를 이용하여 몇 개 그룹으로 클러스터링하는 게 제일 좋은지 판별
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for n_clusters in range_n_clusters:
    model = KMeans(n_clusters = n_clusters,
                   n_init = 10,
                   random_state = 2)
    model.fit(data)
    s_score = silhouette_score(data, model.labels_)
    print('클러스터의 수가 [{}]일 때 실루엘 계수는 [{}]'.format(n_clusters, s_score))
```

# 6절. 연습문제

```python
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
data = iris[['petal_length','petal_width']].values # 독립변수를 array로 안 하면 경고메세지를 출력하는 모형이 있음
```

```python
model = KMeans(n_clusters=2,
              init='random',
              n_init=5).fit(data)
```

```python
# 시각화
centers = model.cluster_centers_ # 중심점 2개
plt.figure(figsize=(12,3))
plt.scatter(x=data[model.labels_==0, 0],
            y=data[model.labels_==0, 1], marker='o', c='b', s=100)
plt.scatter(x=data[model.labels_==1, 0],
            y=data[model.labels_==1, 1], marker='o', c='g', s=100)
plt.scatter(x=centers[:,0], 
            y=centers[:, 1], marker='s', c='r', s=150)
plt.show()
```

```python
# 실루엣 계수
from sklearn.metrics import silhouette_score
silhouette_score(data, model.labels_)
```

```python
# 클러스터 갯수가 3
model3 = KMeans(n_clusters=3,
              init='random',
              n_init=5).fit(data)
```

```python
# 시각화
centers = model3.cluster_centers_ # 중심점 3개
plt.figure(figsize=(12,3))
plt.scatter(x=data[model3.labels_==0, 0],
            y=data[model3.labels_==0, 1], marker='o', c='b', s=100)
plt.scatter(x=data[model3.labels_==1, 0],
            y=data[model3.labels_==1, 1], marker='o', c='g', s=100)
plt.scatter(x=data[model3.labels_==2, 0],
            y=data[model3.labels_==2, 1], marker='^', c='k', s=100)
plt.scatter(x=centers[:,0], 
            y=centers[:, 1], marker='s', c='r', s=150)
plt.show()
```

```python
# 실루엣 계수
from sklearn.metrics import silhouette_score
silhouette_score(data, model3.labels_)
```
