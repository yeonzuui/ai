## ③	결측치를 확인하고 결측치가 없는 컬럼을 포함하여 시각화
## (정렬전 bar plot, barh plot, 정렬 후 bar plot, barh plot)

```python
# 열별 결측치 갯수
missing_cnt = df.isna().sum()
missing_cnt
```

```python
# 결측치를 세로 막대그래프(bar)로 시각화
# missing_cnt.plot(kind = 'bar')
missing_cnt.plot.bar(rot     = 50,
                     figsize = (20, 3))
plt.show()
```

```python
# 결측치를 가로 막대그래프(barh)로 시각화
# missing_cnt.plot(kind = 'barh')
missing_cnt.plot.barh(figsize = (10, 7))
plt.show()
```

```python
missing_cnt_sorted = missing_cnt.sort_values(ascending=False) # 내림차순
missing_cnt_sorted
```

```python
missing_cnt_sorted.plot(kind = 'bar',
                        rot  = 45, 
                        figsize = (15, 3))
plt.show()
```

```python
missing_cnt.sort_values().plot(kind = 'barh', figsize=(7, 7))
plt.show()
```

## ④	결측치가 없는 컬럼은 제외하고 시각화
## (정렬전 bar plot, barh plot, 정렬 후 bar plot, barh plot).

```python
# missing_cnt: 열별 결측치 갯수
only_missing_cnt = missing_cnt[missing_cnt != 0]
only_missing_cnt
```

```python
# only_missing_cnt(결측치가 있는 열 갯수)를 barplot
only_missing_cnt.plot.bar(figsize=(15, 5),
                          rot = 0,
                          ylim = [0, 650000])
for i, v in enumerate(only_missing_cnt):
    if v < 10:
        plt.text(i-.05, v+5000, v)
    elif v < 1000:
        plt.text(i-.1, v+5000, v)
    else:
        plt.text(i-.25, v+5000, v)
plt.show()
```

```python
# only_missing_cnt(결측치가 있는 열 갯수)를 barplot
plt.figure(figsize = (20, 4))
g = sns.barplot(data = only_missing_cnt.to_frame(),
                x = only_missing_cnt.index,
                y = only_missing_cnt,
           )
for i, v in enumerate(only_missing_cnt):
    if v < 10:
        plt.text(i-.05, v+5000, v, color='b', fontsize=12)
    elif v < 1000:
        plt.text(i-.1, v+5000, v, color='b', fontsize=12)
    else:
        plt.text(i-.25, v+5000, v, color='b', fontsize=12)
# plt.title('결측치 갯수')
g.set(title = '결측치 갯수')
g.set_title('결측치 갯수', color='r')
plt.show()
```

```python
only_missing_cnt_sorted = only_missing_cnt.sort_values(ascending=False)
only_missing_cnt_sorted
```

```python
only_missing_cnt_sorted.plot.bar(ylim = [0, 650000],
                                 yticks = [0, 100000, 200000, 400000, 500000, 550000, 600000, 650000],
                                 figsize = (20, 4),
                                 rot = 0
                                )
plt.grid(axis='y', linestyle='--', alpha=.5) # y축만 점선 .5만큼 투명한 점선 grid 추가
for i, v in enumerate(only_missing_cnt_sorted):
    if v < 10:
        plt.text(i-.01, v+5000, v)
    elif v < 100:
        plt.text(i-.05, v+5000, v)
    elif v < 1000:
        plt.text(i-.1, v+5000, v)
    else:
        plt.text(i-.25, v+5000, v, color='r', fontweight='bold', fontsize=12) 
plt.show() 
```

```python
plt.figure(figsize=(20, 7))
g = sns.barplot(data = only_missing_cnt_sorted.to_frame(),
                x = only_missing_cnt_sorted,
                y = only_missing_cnt_sorted.index
               )
g.set_xlim([0, 700000])
g.set_xticks([0, 100000, 200000, 400000, 500000, 550000, 600000, 650000])
plt.grid(axis='x', linestyle='--', alpha=.5) # x축만 점선 .5만큼 투명한 점선 grid 추가
for i, v in enumerate(only_missing_cnt_sorted):
    if v < 1000:
        plt.text(v+5000, i+.2, v)
    else:
        plt.text(v+5000, i+.2, v, color='r', fontweight='bold', fontsize=15) 
plt.show()
```

## ⑤ missingno 라이브러리로 결측치 시각화
* [ResidentMario/missingno: Missing data visualization module for Python](https://github.com/ResidentMario/missingno)

* 위의 사이트를 열어 사용법 확인 ` pip install missingno`

```python
import missingno as msno
msno.matrix(df) # 결측치가 없는 부분은 검정색, 결측치 부분은 흰색
plt.show()
```

```python
msno.bar(df) # 결측치가 아닌 컬럼 갯수를 barplot으로
plt.show()
```

```python
# 결측치가 있는 컬럼에 대해 변수별 상관관계
msno.heatmap(df)
plt.show()
```

```python
# dendrogram: 결측치가 있는 컬럼끼리 군집화
msno.dendrogram(df)
plt.show()
```

# 4. df 데이터 셋에서 사용하지 않을 컬럼은 제거
## ①	결측치가 너무 많은 컬럼은 제거한다. 결측치가 너무 많은 상위 9개 컬럼은 제거함

```python
# 결측치가 많은 컬럼 9개와 결측치 갯수
not_use = df.isnull().sum().sort_values(ascending=False).head(9)
not_use
```

```python
# 호정보 컬럼들의 결측치 비율
not_use['호정보'] / df.shape[0] * 100
```

```python
# not_use 컬럼들의 결측치 비율
round(not_use / df.shape[0]*100, 2)
```

```python
not_use.index
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명       0.02%) - 리스트컴프리헨션
for col, percent in zip(not_use.index, 
                        [str(round(x/len(df)*100, 2)) + '%' for x in not_use]):
    print(col, percent, sep='\t')
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명       0.02%) - map
not_use.map(lambda x: str(round(x/len(df)*100, 2)) + '%')
```

```python
# not_use 컬럼들의 결측치 비율(표준산업분류명       0.02%) - apply
not_use.apply(lambda x: str(round(x/len(df)*100, 2)) + '%')
```

```python
not_use.index # 삭제할 컬럼명
```

```python
df.info()
```

```python
# 컬럼 삭제 전 메모리 사용량(181.5+ MB), 컬럼수
print('컬럼 제거 전 shape:', df.shape)
df.drop(not_use.index, axis=1, inplace=True)
# 컬럼 삭제 후 메모리 사용량(139.6+ MB), 컬럼수
print('컬럼 제거 후 shape:', df.shape)
```

## ②	컬럼명에 “코드”나 “번호”가 있는 컬럼은 분석에 사용되지 않을 예정입니다. 제거함

** 시리즈에 문자함수를 쓰기 위해 참조: https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling

```python
drop_cols
```

```python
print('제거 전 shape:', df.shape)
df.drop(drop_cols, axis=1, inplace=True)
print('제거 후 shape:', df.shape)
```

# 7.	단별량 수치형 변수 시각화
## ①	위도의 빈도표를 시각화: sns.displot, sns.histplot, hist

```python
df[['위도', '경도']].describe()
```

```python
plt.figure(figsize = (12, 3))
df['위도'].hist(bins = 50, 
                grid = False)
plt.show()
```

```python
sns.displot(df['위도'], # figsize 적용 안 됨
            bins = 40,
            rug = True,
            aspect = 2.0)
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.histplot(df['위도'],
             bins = 50
            )
plt.axvline(df['위도'].mean(), 
            color = 'red', 
            linestyle = ':')
plt.annotate('평균',
             xytext = (36, 130000), # 글자 위치
             xy = (df['위도'].mean(), 80000), # 화살표가 가르킬 위치
             arrowprops = dict(arrowstyle = '->',
                               connectionstyle = 'angle3',
                               color = 'r'),
             fontsize = 20,
             color = 'r',
            )
plt.axvline(df['위도'].median(), 
            color ='b', 
            alpha = .5, 
            linestyle = '--')
plt.annotate('중위수',
             xytext = (37.6, 100000), # 글자 위치
             xy = ([df['위도'].median(), 180000]), # 화살표가 가르킬 위치
             arrowprops = dict(arrowstyle = '->',
                               connectionstyle = 'angle',
                               color = 'b',),
             color = 'b',
             fontsize = 15,
            )
plt.show()
```

```python
# 평균을 기준으로 데이터가 치우친 정도
# 음수: 오른쪽으로 치우침 / 양수: 왼쪽으로 치우침
df['위도'].skew()
```
