## 3.3 Seaborn 함수로 그래프 그리기
### 1) 관계형 그래프
    * scatterplot, lineplot, relplot(서브플롯 제공)

```python
# 종별 다른 색상과 다른 스타일의 scatterplot
sns.scatterplot(data=iris,
                x='petal_length',
                y='petal_width',
                hue='species',
                palette='Dark2',
                style='species')
plt.show()
```

```python
# 종별 다른 색상과 스타일의  lineplot
sns.lineplot(data=iris,
             x='petal_length',
             y='petal_width',
             hue='species',
             style='species',
             palette='plasma')
```

```python
# 종별 lineplot을 서브플롯으로 => relplot(관계형 그래프에서 서브플롯 그리는 함수)
sns.relplot(data=iris,
            x='petal_length',
            y='petal_width',
            hue='species',
            kind='line', # 서브플롯으로 그려질 그래프 종류(기본값:scatter)
            col='species', # col 단위로 서브플롯 표현
            # col_wrap=2 # 한 줄에 표현될 서브플롯 수
           )
plt.show()
```

```python
# ax[0]에는 산점도, ax[1]에는 lineplot
fig, ax = plt.subplots(ncols=2)
plt.subplots_adjust(wspace=0.3) # 서브플롯 사이의 공백 조정 
sns.scatterplot(data=iris, 
                x='petal_length', 
                y='petal_width',
                hue='species',
                ax=ax[0])
sns.lineplot(data=iris, 
                x='petal_length', 
                y='petal_width',
                hue='species',
                ax=ax[1])
```

### 2) 범주형 그래프
- stripplot, swamplot, barplot, barhplot, countplot, pointplot
- boxplot, violinplot, boxenplot => catplot(서브플롯)

```python
sns.stripplot(data=iris,
              x='species',
              y='petal_width',
              hue='species' # hue속성이 있을 경우 palette 가능
             )
plt.show()
```

```python
# X축이 범주형일 때의 산점도(중첩되지 않게 )
sns.swarmplot(data=iris,
              x='species',
              y='petal_width',
              hue='species'
             )
plt.show()
```

```python
# 내림차순 정렬된 iris로 범주형 산점도
sns.swarmplot(data=iris.sort_values('petal_width', ascending=False), 
              x='species',
              y='petal_width',
              hue='species'
             )
plt.show()
```

```python
# 산점도의 X축 순서를 임의로 정할 수 있음
sns.swarmplot(data=iris,
              x='species',
              y='petal_width',
              hue='species',
              order=['versicolor', 'virginica', 'setosa']
             )
plt.show()
```

```python
# barplot: 각 범주별 평균값과 그에 대한 신뢰구간
#          신뢰구간 제외: ci(confidence inteval)=None, errorbar=None
sns.barplot(data=iris,
            x='species',
            y='sepal_width',
            # errorbar=None
           )
sns.lineplot(data=iris,
             x='species',
             y='sepal_width'
            )
```

```python
# 성별 servived의 평균을 막대그래프
sns.barplot(data=titanic,
            x='sex',
            y='survived',
            palette='Greens'
           )
```

```python
titanic.info()
```

```python
# class별로 성별 생존률을 하나의 막대그래프
sns.barplot(data=titanic,
            x='sex',
            y='survived',
            hue='class'
           )
plt.show()
```

```python
# class별로 성별 생존률을 서브플롯(막대그래프)
sns.catplot(data=titanic,
            x='sex',
            y='survived',
            hue='class',
            kind='bar', # 기본값은 stripplot
            col='class', # col단위로 subplot 표현
            sharey=False, # y축 공유안함
            # col_wrap=2, # 기본값은 한 행에 모든 서브플롯 표현
           )
```

```python
# boxplot 
sns.boxplot(
            data=titanic,
            x='sex',
            y='age',
            hue='survived'
           )
```
